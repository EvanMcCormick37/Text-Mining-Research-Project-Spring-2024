# Decision Tree Classification

Decision Trees are a popular supervised-learning model for classificiation of data. This means that a decision tree can be used to classify data, but only if it has been first trained on a correctly-labelled training dataset. Decision trees work by finding partition lines along a particular dimension of the data which best reduce the '*impurity*' of the data (*entropy*, or *impurity*, measures how uniform the class labels are within a branch of the decision-tree. For example, a "*pure*" branch would be one in which all data-points are members of the same class. A branch with high *entropy* or *impurity* contains data-points from many different classes.).

These partitions split the larger dataset into smaller groups, or 'branches', which then undergo the same process of subdivision into smaller branches, until each branch of the decision tree contains only points from a single class. At this point, the tree has been built and can be tested for accuracy.

### The Procedure

For this project, I'll be building multiple decision trees to classify the **Ground News** dataset along the three axes of *bias*, *factuality*, and *owner_type*, as outlined in the **Naive Bayes** classification tab. Firstly, I will read in the CSV and convert it into a custom WDM for analysis.

```python
import graphviz
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random as rd
import re
from sklearn import tree
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
#Data
corpus= pd.read_csv("../data/groundnews_corpus_cleaned.csv",index_col=0)
corpus.head()
```

## Raw Data

For this dataset, I'll be creating new WDMs from the dataset using both count-vectorizer and tfidf vectorizer for classification. This will help to determine which of the two methods of vectorization are better for classification.

```python
#Making Vectorizers
MyVect=CountVectorizer(input='content',
                        stop_words='english',
                        token_pattern='(?u)[a-zA-Z]+',
                        lowercase = True
                        )

MyVect_TFIDF=TfidfVectorizer(input='content',
                        analyzer = 'word',
                        stop_words='english',
                        lowercase = True,
                        #binary=True
                        )
```

```python
#Converting the Ground News Data into a WDM
def create_wdm(corpus,index,cv):
    df = pd.DataFrame(cv.fit_transform(corpus[index]).toarray(),columns=cv.get_feature_names_out())
    return df.filter(regex='(?u)[a-zA-Z]{3}+')
count_wdm = create_wdm(corpus,'title_s',MyVect)
tfidf_wdm = create_wdm(corpus,'title_s',MyVect_TFIDF)
wdms=[count_wdm,tfidf_wdm]
bias={'Far Left':-3,'Left':-2,'Lean Left':-1,'Center':0,'Lean Right':1,'Right':2,'Far Right':3}
factuality={'Low Factuality':0,'Mixed Factuality':1,'High Factuality':2}
owner_type={'Other':-1,'Corporation':0,'Government':1,'Independent':2,'Individual':3,'Media Conglomerate':4,'Private Equity':5}
for wdm in wdms:
    wdm['bias']=corpus['bias'].map(bias).fillna(0)
    wdm['factuality']=corpus['factuality'].map(factuality).fillna(-1)
    wdm['owner_type']=corpus['owner_type'].map(owner_type).fillna(1)
```

### Training and Testing Datasets

Because NB is a supervised learning model, we'll need to sort the data into training and testing datasets to avoid over-fitting. To do this, I'll use sklearn's `train_test_split`, with a train:test ratio of $7:3$.

It is especially important to use a train/test split when creating Decision Trees, as they are especially prone to overfitting. Decision trees will always partition the data perfectly into different classes, so long as they contain enough branches. However, these branches often split along random variables which, by coincidence, happen to partition the training data into two or more distinct classes. The testing dataset will help differentiate the random splits from the meaningful ones, and will show if a given decision tree can actually classify the data in general.

```python
#Creating Training and Testing Datasets.
rd.seed(1234)
TrainCount, TestCount = train_test_split(count_wdm, test_size=0.3)
TrainTFIDF, TestTFIDF = train_test_split(tfidf_wdm, test_size=0.3)

TrainCountLabels=TrainCount[['bias','owner_type','factuality']]
TrainTFIDFLabels=TrainTFIDF[['bias','owner_type','factuality']]
TestCountLabels=TestCount[['bias','owner_type','factuality']]
TestTFIDFLabels=TestTFIDF[['bias','owner_type','factuality']]

TrainCount=TrainCount.drop(['bias','owner_type','factuality'],axis=1)
TrainTFIDF=TrainTFIDF.drop(['bias','owner_type','factuality'],axis=1)
TestCount=TestCount.drop(['bias','owner_type','factuality'],axis=1)
TestTFIDF=TestTFIDF.drop(['bias','owner_type','factuality'],axis=1)
```

### Making and Visualizing the Decision Trees

I'll be using sklearn's `DecisionTreeClassifier` for my decision trees, using `entropy` as my impurity measure. It may be useful to constrain the depth/detail of the Decision Tree model to avoid overfitting the data. I don't yet know how these trees will hold up to the testing dataset, so for now I'll allow it to be as deep as it wants to be.

```python
MyDT=DecisionTreeClassifier(criterion='entropy', ##"entropy" or "gini"
                            splitter='best',  ## or "random" or "best"
                            max_depth=None, 
                            min_samples_split=2, 
                            min_samples_leaf=1, 
                            min_weight_fraction_leaf=0.0, 
                            max_features=None, 
                            random_state=None, 
                            max_leaf_nodes=None, 
                            min_impurity_decrease=0.0, 
                            class_weight=None)

matrices=[]

for i in range(3):
    ## perform DT
    #MyDT.fit(TrainDF1, Train1Labels)
    MyDT.fit(TrainCount, TrainCountLabels.iloc[:,i])
    ## plot the tree
    tree.plot_tree(MyDT)
    plt.show()
    feature_names=TrainCount.columns
    dot_data = tree.export_graphviz(MyDT, out_file=None,
                    ## The following creates TrainDF.columns for each
                    ## which are the feature names.
                      feature_names=TrainCount.columns,  
                      #class_names=MyDT.class_names,  
                      filled=True, rounded=True,  
                      special_characters=True)                                    
    graph = graphviz.Source(dot_data) 
    ## Create dynamic graph name
    tempname=str("Graph" + str(i))
    graph.render(tempname)
    ## Show the confusion matrix
    bn_matrix = confusion_matrix(TestCountLabels.iloc[:,i], MyDT.predict(TestCount))
    print("\nThe confusion matrix is:")
    print(bn_matrix)
    matrices+=[bn_matrix]

    ## perform DT
    #MyDT.fit(TrainDF1, Train1Labels)
    MyDT.fit(TrainTFIDF, TrainTFIDFLabels.iloc[:,i])
    ## plot the tree
    tree.plot_tree(MyDT)
    plt.show()
    feature_names=TrainTFIDF.columns
    dot_data = tree.export_graphviz(MyDT, out_file=None,
                    ## The following creates TrainDF.columns for each
                    ## which are the feature names.
                      feature_names=TrainTFIDF.columns,  
                      #class_names=MyDT.class_names,  
                      filled=True, rounded=True,  
                      special_characters=True)                                    
    graph = graphviz.Source(dot_data) 
    ## Create dynamic graph name
    tempname=str("Graph" + str(i))
    graph.render(tempname)
    ## Show the confusion matrix
    bn_matrix = confusion_matrix(TestTFIDFLabels.iloc[:,i], MyDT.predict(TestTFIDF))
    print("\nThe confusion matrix is:")
    print(bn_matrix)
    matrices+=[bn_matrix]
    
```

```python
ConfusionMatrixDisplay(confusion_matrix = matrices[0], 
                                            display_labels = ["Left","Lean Left","Center","Lean Right","Right","Far Right"]).plot()
plt.title("Count, trying to predict Bias")
plt.xticks(rotation=45)
    
ConfusionMatrixDisplay(confusion_matrix = matrices[1], 
                                            display_labels = ["Left","Lean Left","Center","Lean Right","Right","Far Right"]).plot()
plt.title("TFIDF, trying to predict Bias")
plt.xticks(rotation=45)

ConfusionMatrixDisplay(confusion_matrix = matrices[2]
                       #, display_labels = ['Corporation','Government','Independent','Individual','Media Conglomerate','Other','Private Equity']
                      ).plot()
plt.title("Count, trying to predict Owner")
plt.xticks(rotation=45)
    
ConfusionMatrixDisplay(confusion_matrix = matrices[3]
                       #, display_labels = ['Corporation','Government','Independent','Individual','Media Conglomerate','Other','Private Equity']
                      ).plot()
plt.title("TFIDF, trying to predict Owner")
plt.xticks(rotation=45)

ConfusionMatrixDisplay(confusion_matrix = matrices[4]
                       #, display_labels = ['High Factuality', 'Low Factuality', 'Mixed Factuality']
                      ).plot()
plt.title("Count, trying to predict Factuality")
plt.xticks(rotation=45)
    
ConfusionMatrixDisplay(confusion_matrix = matrices[5]
                       #,display_labels = ['High Factuality', 'Low Factuality', 'Mixed Factuality']
                      ).plot()
plt.title("TFIDF, trying to predict Factuality")
plt.xticks(rotation=45)
```

## Takeaways

The decision-tree model doesn't seem to be very effective at classifying the dataset along most of the labelled categories provided, which could be a result of overfitting.

The model performs the best along the '*Factuality*' category, which isn't surprising as that category has relatively few classes and should bear some relation to the content of the text by its nature. 

Next best is the *Bias* category, in which the Decision Tree model clearly underperforms compared to *Naive-Bayes* and *SVM*. It appears that the *Right* class is completely avoided in the decision tree's dest dataset, despite having a population of 14. This could be an artifact of an over-fitted high-depth decision tree. It's indicative that whatever divisions the model used to differentiate the *Right* class from other classes in the training data did not carry over to the testing data, and likely involved superfluous features of the training data.

The *Owner_Type* dataset is the worst, although fortunately the Decision Tree model avoids the '*popular class*' bias present in Naive-Bayes. It does attempt to classify the *Owner_type* correctly into each category. Unfortunately, it fails spectacularly at its job, correctly classifying only $20/105$ classes in the test dataset. That's roughly in-line with random chance in classification power. Combined with the abysmal results of *Naive-Bayes* on this data, there's a strong indication that **owner type** doesn't significantly impact the content of the news articles in this dataset.
